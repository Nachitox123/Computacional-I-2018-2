# Creación de Gráficas

## Datos por Períodos

```python
#Delimitamos los datos de un sólo día.
LimInfDia = '2018-08-31 23:59:59'
LimSupDia = '2018-09-02 00:00:00'

#Delimitamos los datos de una semana.
LimInfSem = LimInfDia
LimSupSem = '2018-09-08 00:00:00'
```

### Canal

```python
#Arreglo para un sólo día
DiaCanal = Canal[(Canal['Fecha'] > LimInfDia) &
                 (Canal['Fecha'] < LimSupDia)]

#Arreglo para una semana
SemCanal = Canal[(Canal['Fecha'] > LimInfSem) &
                 (Canal['Fecha'] < LimSupSem)]
```

### Estación

```python
#Arreglo para un sólo día
DiaEstacion = Estacion[(Estacion['Fecha'] > LimInfDia) &
                       (Estacion['Fecha'] < LimSupDia)]

#Arreglo para una semana
SemEstacion = Estacion[(Estacion['Fecha'] > LimInfSem) &
                       (Estacion['Fecha'] < LimSupSem)]

SemEstacion = SemEstacion.drop(SemEstacion.index[671])
```

### Salinidad

```python
#Arreglo para un sólo día
DiaSalinidad = Salinidad[(Salinidad['Fecha'] > LimInfDia) &
                         (Salinidad['Fecha'] < LimSupDia)]

#Arreglo para una semana
SemSalinidad = Salinidad[(Salinidad['Fecha'] > LimInfSem) &
                         (Salinidad['Fecha'] < LimSupSem)]

SemSalinidad = SemSalinidad.drop(SemSalinidad.index[671])
```

## Gráficas

### Nivel del Mar

#### Nivel del mar en la ESTACIÓN y el CANAL por 1 DÍA.

```python
#Datos a graficar
x = DiaEstacion['Fecha']
y1 = DiaEstacion['NivelAgua']
y2 = DiaCanal['NivelAgua']

plt.title('Nivel del Agua (1 Día)', fontweight='bold')  #Título
plt.xlabel('Hora', fontweight='bold')                  #Eje X
plt.xticks(rotation=45)                                 #Datos del eje X

#plt.xlim(['2018-09-01 00:00:00','2018-09-01 23:59:59'])

EjeIzq = plt.subplot()    #Creamos el primer eje, tomando en cuenta que serán dos.
EjeDer = EjeIzq.twinx()   #Creamos el segundo, indicando que comparten X.

#Primer eje --------------------------------------------------
EjeIzq.set_ylabel('Estación (m)', color = 'b', fontweight='bold')
EjeIzq.tick_params(axis='y', labelcolor = 'b')
EjeIzq.plot(x,y1, color='b')

#Segundo eje -------------------------------------------------
EjeDer.set_ylabel('Canal (m)', color='r', fontweight='bold')
EjeDer.tick_params(axis='y', labelcolor= 'r')
EjeDer.plot(x, y2, color='r')

plt.grid()
plt.tight_layout()
plt.savefig('Nivel del Agua - Dia')
```

### Correlación

#### Correlación entre el NIVEL DEL AGUA de la ESTACIÓN y la SALINIDAD para 1 DÍA.

```python
#Crearemos un nuevo DataFrame con los datos del NIVEL DEL AGUA de la ESTACIÓN.
DatosCorr = DiaEstacion.drop('Fecha', axis=1)

#Las columnas restantes son información de la SALINIDAD.
DatosCorr['ConduEspec'] = DiaSalinidad.ConduEspec
DatosCorr['Salinidad'] = DiaSalinidad.Salinidad

#Eliminamos aquellos renglones donde exista un NaN.
DatosCorr = DatosCorr.dropna()

#Calculos la correlación entre las variables
Corre = DatosCorr.corr()

#Creamos el mapa de calor.
Mapa = sns.heatmap(Corre, square=True, annot=True, annot_kws={'size': 12}, linewidths= 1,
                   linecolor='black', cmap='YlGnBu')

Imagen = Mapa.get_figure()
Imagen.tight_layout()
Imagen.savefig('Correlacion Agua y Salinidad - 1 Dia')
```

#### Correlación entre el NIVEL DEL AGUA de la ESTACIÓN y la SALINIDAD para 1 SEMANA.

```python
#Crearemos un nuevo DataFrame con los datos del NIVEL DEL AGUA de la ESTACIÓN.
DatosCorr = SemEstacion.drop('Fecha', axis=1)

#Las columnas restantes son información de la SALINIDAD.
DatosCorr['ConduEspec'] = SemSalinidad.ConduEspec
DatosCorr['Salinidad'] = SemSalinidad.Salinidad

#Eliminamos aquellos renglones donde exista un NaN.
DatosCorr = DatosCorr.dropna()

#Calculos la correlación entre las variables
Corre = DatosCorr.corr()

#Creamos el mapa de calor.
Mapa = sns.heatmap(Corre, square=True, annot=True, annot_kws={'size': 12}, linewidths= 1,
            linecolor='black', cmap='YlGnBu')

Imagen = Mapa.get_figure()
Imagen.tight_layout()
Imagen.savefig('Correlacion Agua y Salinidad - 1 Semana')
```
