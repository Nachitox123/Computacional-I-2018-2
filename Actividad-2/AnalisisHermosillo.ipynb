```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Lectura del archivo (direcci'on, separador, salto de lineas, encabezado)
Estacion = pd.read_csv('UNISON_90.csv', skiprows=9, header=None)
```

```python
Estacion
```

```python
#Asignaremos los nombres de cada columna
Estacion.columns = ['Fecha', 'DirRaf', 'DirVien', 'HumRel', 'Precip',
                    'RadSolar', 'Tempe', 'PresAtm']
```

```python
Estacion.head()
```

```python
Estacion.dtypes
```

```python
#La columna FECHA se marca como OBJECT, ahora lo convertiremos a DATETIME
Estacion['Fecha'] = pd.to_datetime(Estacion['Fecha'])
```

```python
#Ahora invertiremos los datos para tener de manera ascendentes las fechas
Esta2 = Estacion[::-1]
```

```python
#Iniciamos el conteo de las filas nuevamente
Esta2.reset_index(drop=True)
```

```python
#Solamente necesitamos los 'ultimos 10 dias, debemos seleccionarlos.
DiezDias = Esta2[(Esta2['Fecha'] > '2018-08-24 23:59:59') &
               (Esta2['Fecha'] < '2018-09-04 00:00:00')]

print(DiezDias)
```

```python
#Gr'afica de Temperatura durante los 'ultimos 10 d'ias
plt.title('Temperatura de 10 días', fontweight='bold')

plt.xlabel('Días', fontweight='bold')
plt.xticks(rotation=45)
plt.ylabel('Temperatura (°C)', fontweight='bold')

plt.grid(True)

x = DiezDias['Fecha']
y = DiezDias['Tempe']

plt.plot(x,y, 'b')

plt.savefig('Tempe10', bbox_inches='tight')
```

```python
#Gr'afica de Presi'on durante los 'ultimos 10 d'ias
plt.title('Presión de 10 días', fontweight='bold')

plt.xlabel('Días', fontweight='bold')
plt.xticks(rotation=45)
plt.ylabel('Presión (hPa)', fontweight='bold')

plt.grid(True)

x = DiezDias['Fecha']
y = DiezDias['PresAtm']

plt.plot(x,y, 'g')

plt.savefig('Pres10', bbox_inches='tight')
```
